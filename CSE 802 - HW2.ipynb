# **Q6**

Consider the following class-conditional densities for a three-class problem involving two-
dimensional features:
p(x|ω1) ∼ N 􏰀(−1,−1)t, I􏰁; p(x|ω2) ∼ N 􏰀(1,1)t, I􏰁;
p(x|ω3)∼ 12N􏰀(0.5,0.5)t,I􏰁+ 12N􏰀(−0.5,−0.5)t,I􏰁.

(a) In a 2D graph, mark the mean of ω1, ω2, and the two components of ω3. In the same graph, mark the point x = (0.1, 0.1)t .

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.stats import multivariate_normal

# Define the class-conditional densities
p1 = multivariate_normal(mean=[-1, -1], cov=np.eye(2))
p2 = multivariate_normal(mean=[1, 1], cov=np.eye(2))
p3_1 = multivariate_normal(mean=[0.5, 0.5], cov=np.eye(2))
p3_2 = multivariate_normal(mean=[-0.5, -0.5], cov=np.eye(2))

# Plot the class means and the point (0.1, 0.1)
fig, ax = plt.subplots()
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y

# plot the ellipses for the Gaussian Mixture Model
for i, cov in enumerate([np.eye(2), np.eye(2)]):
    lambda_, v = np.linalg.eig(cov)
    lambda_ = np.sqrt(lambda_)
    ell = Ellipse(xy=[0, 0], width=lambda_[0]*3, height=lambda_[1]*3, angle=np.rad2deg(np.arccos(v[0, 0])), facecolor='none', edgecolor='r')
    ell.set_clip_box(ax.bbox)
    ell.set_edgecolor('r')
    if i == 0: #for p3_1 and p3_2
        ell.set_center([0.5, 0.5])
    else:
        ell.set_center([-0.5, -0.5])
    ax.add_artist(ell)

# plot the means and the point
ax.contour(x, y, p1.pdf(pos), colors='b')
ax.contour(x, y, p2.pdf(pos), colors='g')
ax.contour(x, y, p3_1.pdf(pos) + p3_2.pdf(pos), colors='r')

#plot x = (0.1,0.1)
ax.scatter(0.1, 0.1, marker='x', color='k', s=100)

#plot p1
plt.scatter(-1, -1, marker='o', color='b', s=100)

#plot p2
plt.scatter(1, 1, marker='o', color='g', s=100)

#plot p3_1
plt.scatter(0.5, 0.5, marker='o', color='r', s=100)
#plot p3_2
plt.scatter(-0.5, -0.5, marker='o', color='r', s=100)

plt.xlabel('x1')
plt.ylabel('x2')
plt.show()

```

# **Q8**

a) Train the classifier: Using the first 25 patterns of each class (training data), compute μj and Σj, j = 1, 2, 3. Report these values.

(b) Design the Bayes classifier: Assuming that the three classes are equally probable and a 0-1 loss function, write a program that inputs a 4-dimensional pattern x and assigns it to one of the three classes
j∗ = arg max {P(ωj|x)}. j=1,2,3

(c) Test the classifier: Classify the remaining 25 patterns of each class (test data) using the Bayes classifier constructed above and report the confusion matrix for this three-class problem. What is the empirical error rate on the test set?

```python
import numpy as np
from sklearn.datasets import load_iris
from scipy.stats import multivariate_normal

# Load the iris dataset
iris = load_iris()
# Extract the first 25 patterns from each class for training
train_data = np.vstack((iris.data[:25], iris.data[50:75], iris.data[100:125]))

# Compute the mean and covariance matrix for each class
mu = [np.mean(train_data[:25], axis=0), np.mean(train_data[25:50], axis=0), np.mean(train_data[50:75], axis=0)]
cov = [np.cov(train_data[:25], rowvar=False), np.cov(train_data[25:50], rowvar=False), np.cov(train_data[50:75], rowvar=False)]

# Classify the remaining 25 patterns from each class as test data
test_data = np.vstack((iris.data[25:50], iris.data[75:100], iris.data[125:150]))

# Initialize confusion matrix
confusion_matrix = np.zeros((3, 3))

# Iterate over test data and classify each pattern using the Bayes classifier
for i in range(len(test_data)):
    x = test_data[i] #4-dimensional pattern x input
    posteriors = []
    for j in range(3):
        posterior = multivariate_normal(mu[j], cov[j]).pdf(x) * (1/3)
        posteriors.append(posterior)
    predicted_class = np.argmax(posteriors) #maximum posterior rule argmax
    true_class = int(i / 25)
    confusion_matrix[true_class][predicted_class] += 1

# Compute empirical error rate
empirical_error_rate = 1 - np.trace(confusion_matrix) / np.sum(confusion_matrix)

# Print the results
print("Mean for each class:")
print('setosa')
print(mu[0])
print('versicolor')
print(mu[1])
print('virginica')
print(mu[2])
print("Covariance matrix for each class:")
print('setosa')
print(cov[0])
print('versicolor')
print(cov[1])
print('virginica')
print(cov[2])
print("Confusion matrix:")
print(confusion_matrix)
print("Empirical error rate:", empirical_error_rate)
```
